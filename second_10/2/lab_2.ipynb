2. Проведите первичный анализ данных. Особое внимание следует уделить графическому представлению распрееделений признаков, визуализации взаимосвзей, позволяющие судить о наборе данных. Построение графиков желательно произвести по нескольким проекциям. При анализе данных использовать как можно более разнообразные типы графиков.

```python
# Импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_path = 'car.data'  # Замените на ваш путь к файлу
data = pd.read_csv(file_path, header=None)

# Переименование столбцов
columns = ['Buying', 'Maint', 'Doors', 'Persons', 'Lug_Boot', 'Safety', 'Class']
data.columns = columns

# Исследование данных
print(data.info())
print(data.head())

# Распределение значений для каждого признака
sns.set(style="whitegrid")
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()
for i, col in enumerate(data.columns):
    sns.countplot(x=data[col], ax=axes[i], palette="Set2")
    axes[i].set_title(f"Распределение {col}")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

# Взаимосвязь между признаками
pairs = [
    ('Buying', 'Class'),
    ('Maint', 'Class'),
    ('Doors', 'Class'),
    ('Persons', 'Class'),
    ('Lug_Boot', 'Safety'),
    ('Safety', 'Class'),
]

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()
for i, col in enumerate(data.columns):
    sns.countplot(x=data[col], ax=axes[i], color="skyblue")  # Убрали palette
    axes[i].set_title(f"Распределение {col}")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

**Выводы по графикам**

Распределение признаков:
1. Buying и Maint: Распределение равномерное, все категории (vhigh, high, med, low) представлены примерно одинаково.
2. Doors: Категория 5more имеет немного меньшее количество записей, чем другие.
3. Persons: Чаще всего встречаются значения 2 и 4, категория more представлена реже.
4. Lug_Boot: Размеры багажников (small, med, big) распределены примерно равномерно.
5. Safety: Категория low встречается немного реже, чем med и high.
6. Class: Категория unacc преобладает, что может указывать на сильный дисбаланс данных.

Взаимосвязи между признаками:
1. Buying и Class: Класс unacc преобладает во всех категориях стоимости покупки, что указывает на ограниченную зависимость от этого признака.
2. Maint и Class: Аналогично, unacc доминирует, но в категориях low и med чаще встречаются классы acc и vgood.
3. Doors и Class: Признак Doors мало влияет на распределение классов.
4. Persons и Class: Категория more способствует увеличению доли классов acc и vgood, что может быть важным признаком.
5. Lug_Boot и Safety: Более высокий уровень безопасности (med, high) чаще встречается с более крупным багажником.
6. Safety и Class: Высокий уровень безопасности (high) связан с увеличением доли классов acc, good и vgood.
