**1.Какие числовые признаки называются дискретными, а какие — непрерывными? Приведите собственные примеры.**

Дискретные числовые признаки — это признаки, которые принимают только отдельные значения, обычно их можно пересчитать. Пример: количество студентов в группе.

Непрерывные числовые признаки — это признаки, которые могут принимать любые значения в пределах некоторого диапазона, их можно измерить. Пример: температура воды в озере.

**2.Опишите разницу между обучением с учителем и обучением без учителя.**

Обучение с учителем предполагает использование набора данных, где каждому экземпляру соответствуют как входные данные, так и правильный ответ. Алгоритм учится на этих примерах, чтобы предсказывать результаты на новых данных.

Обучение без учителя осуществляется без заранее известных ответов. Алгоритму предоставляются только данные, и его задача — самостоятельно выявить скрытые структуры или закономерности.

**3.Опишите разницу между задачами классификации и регрессии.**

Задача классификации заключается в том, чтобы предсказать категорию или класс объекта, разделяя данные на несколько групп. Обычно используется в обучении с учителем.

Задача регрессии решается, когда нужно предсказать непрерывное значение, которое может принимать любые числовые значения. (Y ∈ R)

**5.Почему в качестве альтернативы простому использованию всех тренировочных данных рекомендуется выполнять проверку модели на валидационных данных?**

Проверка модели на валидационных данных помогает избежать переобучения. Мы можем выделить часть тренировочных данных, обучить модель на оставшейся части, а потом проверить её на отложенных данных. Этот метод называется валидацией с использованием отложенной выборки.

**6.Даны три матрицы A, B, C. Матрица A имеет размеры 5 × 4, B — размеры 4 × 6, C имеет размеры 3 × 5. Укажите все возможные матрицы, которые можно перемножить между собой.**

Можно перемножить матрицы A×B (5×4 и 4×6), так как число столбцов A совпадает с числом строк B.
Другие комбинации не перемножаются, так как их размеры не соответствуют правилам матричного умножения.

**7.Найдите градиент функции **

Градиент функции - это вектор, состоящий из частных производных функции по каждой переменной.

Частная производная по x: ∂f/∂x = 2yx

Частная производная по y: ∂f/∂y = x^2 + 1/y

Частная производная по z: ∂f/∂z = -e^(-z)

Таким образом, градиент функции будет равен: f(x, y, z) = (2yx, x^2 + 1/y, -e^(-z))
